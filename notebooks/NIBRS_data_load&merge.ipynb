```python
import pandas as pd
import os

def load_nibrs_folder(folder_path):

    incident = pd.read_csv(
        os.path.join(folder_path, "NIBRS_incident.csv"),
        encoding="latin1",
        low_memory=False
    )

    offense = pd.read_csv(
        os.path.join(folder_path, "NIBRS_OFFENSE.csv"),
        encoding="latin1",
        low_memory=False
    )

    offense_type = pd.read_csv(
        os.path.join(folder_path, "NIBRS_OFFENSE_TYPE.csv"),
        encoding="latin1",
        low_memory=False
    )

    agencies = pd.read_csv(
        os.path.join(folder_path, "agencies.csv"),
        encoding="latin1",
        low_memory=False
    )

    location = pd.read_csv(
        os.path.join(folder_path, "NIBRS_LOCATION_TYPE.csv"),
        encoding="latin1",
        low_memory=False
    )

    # merge 
    df = incident.merge(agencies, on="agency_id", how="left")
    df = df.merge(offense, on="incident_id", how="left")
    df = df.merge(offense_type, on="offense_code", how="left")
    df = df.merge(location, on="location_id", how="left")

    return df
```

```python
base_dir = r"D:\nuscourses\IT5006\NIBRS DATASET"

folders = [
    "CA-2021", "CA-2022", "CA-2023", "CA-2024",
    "TX-2021", "TX-2022", "TX-2023", "TX-2024",
    "NY-2021", "NY-2022", "NY-2023", "NY-2024"
]

all_dfs = []

for folder in folders:
    folder_path = os.path.join(base_dir, folder)
    df_temp = load_nibrs_folder(folder_path)
    df_temp["source_folder"] = folder
    all_dfs.append(df_temp)

df_all = pd.concat(all_dfs, ignore_index=True)
```
