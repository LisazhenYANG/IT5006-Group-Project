```python
import numpy as np
import pandas as pd

np.random.seed(42)

n_samples = 1000

feature_cols = [
    "commercial_ratio",
    "institution_ratio",
    "other_ratio",
    "public_ratio",
    "residential_ratio",
    "theft_ratio",
    "battery_ratio",
    "criminal_damage_ratio",
    "assault_ratio",
    "deceptive_practice_ratio",
    "other_crime_ratio"
]

# Generate random ratios (sum to 1 using Dirichlet distribution)
X = np.random.dirichlet(np.ones(len(feature_cols)), size=n_samples)

df = pd.DataFrame(X, columns=feature_cols)

# Generate a somewhat realistic hotspot label
# (weighted combination instead of pure random)
weights = np.array([0.5, 0.3, 0.2, 0.4, 0.1, 0.6, 0.5, 0.4, 0.3, 0.2, 0.2])
score = df.values @ weights

threshold = np.percentile(score, 70)  # top 30% as hotspot
df["hotspot"] = (score > threshold).astype(int)

# Save to data folder
df.to_csv("../data/apptest.csv", index=False)

print("apptest.csv created successfully!")
print(df.head())
```
